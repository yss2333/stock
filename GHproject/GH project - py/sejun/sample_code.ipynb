```python
stocks =  pd.read_csv('C:\\Users\\yss06\\Desktop\\python\\stock data.csv', header=0, encoding="utf-8-sig")

stocks_normalized = stocks.copy()
columns_to_normalize = stocks.columns.difference(['Date'])

# Apply Min-Max normalization to the specified columns
for col in columns_to_normalize:
    stocks_normalized[col] = (stocks_normalized[col] - stocks_normalized[col].min()) / (stocks_normalized[col].max() - stocks_normalized[col].min())
stocks_normalized
```

```python
TEST_SIZE = 200
WINDOW_SIZE = 20

train=stocks_normalized[:-TEST_SIZE]
test = stocks_normalized[-TEST_SIZE:]
```

```python
def make_dataset(stocks_normalized, label, window_size=20):
    feature_list = []
    label_list = []
    for i in range(len(stocks_normalized) - window_size):
        feature_list.append(np.array(stocks_normalized.iloc[i:i+window_size]))
        label_list.append(np.array(label.iloc[i+window_size]))
    return np.array(feature_list), np.array(label_list)
```

```python
from sklearn.model_selection import train_test_split

feature_cols = ['Open', 'High', 'Low','Volume','Change']
label_cols = ['Close']

train_feature = train[feature_cols]
train_label = train[label_cols]

train_feature, train_label = make_dataset(train_feature, train_label, 20)

x_train, x_valid, y_train, y_valid = train_test_split(train_feature, train_label, test_size=0.2)
x_train.shape, x_valid.shape
```

```python
test_feature = test[feature_cols]
test_label = test[label_cols]

test_feature.shape, test_label.shape
```

```python
test_feature, test_label = make_dataset(test_feature, test_label, 20)
test_feature.shape, test_label.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping, ModelCheckpoint
from keras.layers import LSTM

model = Sequential()
model.add(LSTM(16, 
               input_shape=(train_feature.shape[1], train_feature.shape[2]), 
               activation='relu', 
               return_sequences=False)
          )

model.add(Dense(1))
```

```python
import os

model.compile(loss='mean_squared_error', optimizer='adam')
early_stop = EarlyStopping(monitor='val_loss', patience=5)

model_path = 'model'
filename = os.path.join(model_path, 'tmp_checkpoint.h5')
checkpoint = ModelCheckpoint(filename, monitor='val_loss', verbose=1, save_best_only=True, mode='auto')

history = model.fit(x_train, y_train, 
                                    epochs=200, 
                                    batch_size=16,
                                    validation_data=(x_valid, y_valid), 
                                    callbacks=[early_stop, checkpoint])
```

```python
model.load_weights(filename)
pred = model.predict(test_feature)

pred.shape
```

```python
plt.figure(figsize=(12, 9))
plt.plot(test_label, label = 'actual')
plt.plot(pred, label = 'prediction')
plt.legend()
plt.show()
```
